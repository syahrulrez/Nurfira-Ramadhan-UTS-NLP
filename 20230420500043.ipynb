1. Text Cleaning (Pembersihan Teks)

```python
import re

def clean_text(text):
    text = str(text)
    text = re.sub(r'http\S+', '', text)  # hapus URL
    text = re.sub(r'[^a-zA-Z\s]', '', text)  # hapus angka & simbol
    text = re.sub(r'\s+', ' ', text).strip()  # hapus spasi berlebih
    return text

# Fix for the KeyError: 'Berita' due to incorrect CSV parsing
# If 'No;Berita' column exists, extract the 'Berita' part.
if 'No;Berita' in df.columns:
    # Use str.partition to safely get the part after the first semicolon
    # The partition method returns a DataFrame, so we access the third column directly by index [2]
    df['Berita_content'] = df['No;Berita'].astype(str).str.partition(';')[2]
    df['cleaning'] = df['Berita_content'].apply(clean_text)
    df[['Berita_content', 'cleaning']].head()
elif 'Berita' in df.columns:
    # If 'Berita' column already exists (e.g., if the CSV reading was fixed elsewhere),
    # proceed with the original logic.
    df['cleaning'] = df['Berita'].apply(clean_text)
    df[['Berita', 'cleaning']].head()
else:
    # Handle case where neither expected column is found (e.g., different column names)
    print("Error: Neither 'Berita' nor 'No;Berita' column found. Please check your DataFrame columns.")
    print("Available columns:", df.columns.tolist())
```

```python
import re
import string
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

# Pastikan sudah mendownload resource NLTK yang diperlukan
import nltk
nltk.download('punkt')
nltk.download('stopwords')

def clean_text(text):
    # 1. Mengubah teks menjadi lowercase (huruf kecil)
    text = text.lower()

    # 2. Menghapus URL (jika ada)
    text = re.sub(r'http\S+|www\S+|https\S+', '', text, flags=re.MULTILINE)

    # 3. Menghapus karakter spesial, angka, dan emoji
    # Kita hanya menyisakan huruf a-z
    text = re.sub(r'[^a-z\s]', '', text)

    # 4. Menghapus tanda baca (punctuation)
    text = text.translate(str.maketrans('', '', string.punctuation))

    # 5. Tokenisasi (memecah kalimat menjadi kata)
    tokens = word_tokenize(text)

    # 6. Menghapus Stopwords (kata umum yang kurang bermakna seperti 'yang', 'di', 'ke')
    stop_words = set(stopwords.words('indonesian'))
    # Opsional: Tambahkan kata spesifik yang ingin dibuang (misal: 'kompascom')
    additional_stop_words = {'kompascom', 'baca', 'juga'}
    stop_words.update(additional_stop_words)

    filtered_text = [word for word in tokens if word not in stop_words]

    # Menggabungkan kembali menjadi satu string
    return " ".join(filtered_text)

# Input Kalimat
raw_text = """
JAKARTA, KOMPAS.com - Komisi Pemberantasan Korupsi (KPK) mengatakan, nama pengusaha rokok Khairul Umam alias Haji Her ada tercantum dalam dokumen yang ditemukan penyidik saat menggeledah kantor Direktorat Jenderal Bea Cukai (DJBC). Plt Direktur Penyidikan KPK Achmad Taufik Husein mengatakan, termuat tersebut menjadi alasan KPK memanggil Haji Her sebagai saksi terkait kasus dugaan korupsi di lingkungan DJBC. “Haji Her, saya kembali, jadi hasil penggeledahan yang kita temukan di proses penyidikan di kantor Ditjen Bea Cukai itu ada beberapa dokumen-dokumen yang dibuat oleh Ocoy (Kasi Intelijen DJBC Orlando Hamonangan) si tersangka ini,” kata Taufik, di Gedung Merah Putih, Jakarta, Senin (13/4/2026). “Kemudian kita analisis-analisis. Di situlah ditemukan beberapa nama-nama pengusaha rokok. Suryo termasuk Haji Her,” sambung dia. Baca juga: Tersangka Siman Bahar Meninggal Dunia, KPK Bakal Siapkan SP3 :) !!!
"""

cleaned_text = clean_text(raw_text)

print("--- Teks Sebelum Cleaning ---")
print(raw_text)
print("\n--- Teks Sesudah Cleaning ---")
print(cleaned_text)
```

```python
import re

def normalize_text(text):
    # Mapping normalisasi: "Kata Tidak Baku/Singkatan": "Kata Baku"
    normalization_dict = {
        "kpk": "komisi pemberantasan korupsi",
        "djbc": "direktorat jenderal bea cukai",
        "plt": "pelaksana tugas",
        "ditjen": "direktorat jenderal",
        "sp3": "surat perintah penghentian penyidikan",
        "kasi": "kepala seksi"
    }

    # 1. Kecilkan huruf terlebih dahulu agar match dengan dictionary
    text = text.lower()

    # 2. Tokenisasi sederhana untuk mencocokkan kata per kata
    words = text.split()

    # 3. Proses penggantian kata
    normalized_words = []
    for word in words:
        # Menghapus tanda baca yang menempel pada kata (misal: "djbc.")
        clean_word = word.strip(string.punctuation)

        if clean_word in normalization_dict:
            # Ganti dengan versi panjangnya
            replacement = normalization_dict[clean_word]
            normalized_words.append(replacement)
        else:
            normalized_words.append(word)

    return " ".join(normalized_words)

# Contoh penggunaan dengan teks yang sudah dibersihkan sedikit (atau teks asli)
raw_text = "Plt Direktur Penyidikan KPK mengatakan ada dokumen di kantor DJBC terkait tersangka."

normalized_result = normalize_text(raw_text)

print("--- Sebelum Normalisasi ---")
print(raw_text)
print("\n--- Sesudah Normalisasi ---")
print(normalized_result)
```

```python
from nltk.tokenize import word_tokenize
from nltk import pos_tag
from nltk.stem import WordNetLemmatizer
from nltk.corpus import wordnet # Import wordnet for POS tags
import nltk
nltk.download('averaged_perceptron_tagger')
nltk.download('averaged_perceptron_tagger_eng')

lemmatizer = WordNetLemmatizer()
sentence = "JAKARTA, KOMPAS.com - Komisi Pemberantasan Korupsi (KPK) mengatakan, nama pengusaha rokok Khairul Umam alias Haji Her ada tercantum dalam dokumen yang ditemukan penyidik saat menggeledah kantor Direktorat Jenderal Bea Cukai (DJBC). Plt Direktur Penyidikan KPK Achmad Taufik Husein mengatakan, termuat tersebut menjadi alasan KPK memanggil Haji Her sebagai saksi terkait kasus dugaan korupsi di lingkungan DJBC. “Haji Her, saya kembali, jadi hasil penggeledahan yang kita temukan di proses penyidikan di kantor Ditjen Bea Cukai itu ada beberapa dokumen-dokumen yang dibuat oleh Ocoy (Kasi Intelijen DJBC Orlando Hamonangan) si tersangka ini,” kata Taufik, di Gedung Merah Putih, Jakarta, Senin (13/4/2026). “Kemudian kita analisis-analisis. Di situlah ditemukan beberapa nama-nama pengusaha rokok. Suryo termasuk Haji Her,” sambung dia. Baca juga: Tersangka Siman Bahar Meninggal Dunia, KPK Bakal Siapkan SP3 :) !!!"
tokens = word_tokenize(sentence)
tagged_tokens = pos_tag(tokens)

def get_wordnet_pos(tag):
    if tag.startswith('J'):
        return 'a'
    elif tag.startswith('V'):
        return 'v'
    elif tag.startswith('N'):
        return 'n'
    elif tag.startswith('R'):
        return 'r'
    else:
        return 'n'

lemmatized_sentence = []
for word, tag in tagged_tokens:
    lemmatized_sentence.append(lemmatizer.lemmatize(word, get_wordnet_pos(tag)))

print(f"Original Sentence: {sentence}")
print(f"Lemmatized Sentence: {' '.join(lemmatized_sentence)}")
```
